```python
def setup():
    size(600,600)
    rect_mode(CENTER)

def draw():
    
    push_matrix() 
    translate(width/2, width/2)
    background(255)
    p = 1
    for i in range(3, 20):
        rotate(p*PI/i)
        no_fill()
        stroke_weight(2)
        rect(0,0,i*30,i*30)

    pop_matrix()
```
