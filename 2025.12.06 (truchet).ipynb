### PADR√ÉO DE TRUCHET

(https://en.wikipedia.org/wiki/Truchet_tile)

```python
raio = 5
def setup():
    size(1200, 800)
    background(255)
    p = random_int(3)
    for i in range(int(width/raio)):
        for j in range (int(height/raio)): 
            p = random_int(1)

            if p == 0:
                stroke_weight(2)
                stroke(2*i,2*j,0)
                line(raio*i, raio*j, raio*i+raio ,raio*j + raio )
            else:
                stroke(2*i,2*j,0)
                stroke_weight(2)
                line(raio*i+raio, raio*j, raio*i, raio*j+raio) 

```
